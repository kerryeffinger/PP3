```python
import os
import csv
import numpy as np
from prettytable import PrettyTable
import statistics as stat
```

## 1.

```python
def read_africa_data(filename):
    """this function accepts a csv file and returns the data as a list"""
    with open (filename, mode = 'r', newline='') as ap:
        reader = csv.reader(ap)
        africalist= []
        for record in reader:
            africalist.append(record)
    return africalist
```

```python
africal= read_africa_data('africa_pop_data.csv')
```

## 2.

```python
def array_africa_data(dataset):
    """this function accepts a dataset and makes a NumPy array from the numerical data"""
    numericalafrica = [item[2:7] for item in dataset]
    array = np.array(numericalafrica) 
    
    return array[1:,:]
```

## 3.

```python
def pt_africa_data(dataset):
    """this function accepts a dataset and returns a prettytable table of the data"""
    header = dataset[0]
    p = PrettyTable(header)
    for row in africal:
        p.add_row(row)
    p.del_row(0)  
    p.align['Country'] = "r"
    p.align['2015'] ='r'
    p.align['2016'] ='r'
    p.align['2017'] ='r'
    p.align['2018'] ='r'
    p.align['2019'] ='r'
    print(p)
```

## 4.

```python
def arrayint(array):
    """this function formats data from a numpy array as integers and finds the min, max and mean of each column"""
    int2015 = array[:,0:1].astype(np.int)
    int2016 = array[:,1:2].astype(np.int)
    int2017 = array[:,2:3].astype(np.int)
    int2018 = array[:,3:4].astype(np.int)
    int2019 = array[:,4:].astype(np.int)
    mins = np.min(int2015), np.min(int2016), np.min(int2017), np.min(int2018), np.min(int2019)
    maxs = np.max(int2015), np.max(int2016), np.max(int2017), np.min(int2018), np.max(int2019)
    means = np.mean(int2015), np.mean(int2016), np.mean(int2017), np.mean(int2018), np.mean(int2019)

    return mins, maxs, means

def summary_africa(array):
    """this function accepts a numpy array and returns a table with summary statistics of the data"""
    mins, maxs, means = arrayint(array)
    header = ['Summary Measure','2015','2016','2017','2018','2019']
    p = PrettyTable(header)
    print(p)
    print(mins,maxs,means)
```
